# Antrian

```python
from collections import deque # Mengimpor kelas deque dari modul collections
queue = deque(["Ram", "Tarun", "Asif", "John"]) # Membuat deque dengan nama queue yang berisi elemen-elemen awal "Ram", "Tarun", "Asif", dan "John"
print(queue) # Mencetak isi deque pada saat itu. Hasilnya akan menjadi deque(['Ram', 'Tarun', 'Asif', 'John'])
queue.append("Akbar") # Menambahkan string "Akbar" ke ujung kanan deque
print(queue) # Mencetak isi deque setelah menambahkan "Akbar". Hasilnya akan menjadi deque(['Ram', 'Tarun', 'Asif', 'John', 'Akbar'])
queue.append("Birbal") # Menambahkan string "Birbal" ke ujung kanan deque
print(queue) # Mencetak isi deque setelah menambahkan "Birbal". Hasilnya akan menjadi deque(['Ram', 'Tarun', 'Asif', 'John', 'Akbar', 'Birbal'])
print(queue.popleft()) # Menghapus dan mengembalikan elemen paling kiri dari deque. Pada contoh ini, menghapus "Ram" dan mencetaknya. Hasilnya akan menjadi Ram
print(queue.popleft()) # Menghapus dan mengembalikan elemen paling kiri dari deque. Kali ini, menghapus "Tarun" dan mencetaknya. Hasilnya akan menjadi Tarun
print(queue) # Mencetak isi deque setelah dua elemen paling kiri dihapus. Hasilnya akan menjadi deque(['Asif', 'John', 'Akbar', 'Birbal'])
```

# Tower Hanoi

```python
# Mendefinisikan fungsi
def TowerOfHanoi(n, from_rod, to_rod, aux_rod):
  # n: Jumlah cakram yang akan dipindahkan.
  # from_rod: Nama batang awal (rod) dari mana cakram akan dipindahkan.
  # to_rod: Nama batang tujuan (rod) ke mana cakram akan dipindahkan.
  # aux_rod: Nama batang bantu (auxiliary rod).
    if n == 0: #  Ini adalah kondisi dasar dari rekursi. Jika jumlah cakram (n) sudah nol, maka fungsi akan langsung mengembalikan tanpa melakukan apa pun
        return
    TowerOfHanoi(n-1, from_rod, aux_rod, to_rod) # Pemanggilan rekursif pertama untuk memindahkan (n-1) cakram dari batang awal ke batang bantu
    print("Move disk", n, "from rod", from_rod, "to rod", to_rod) # Mencetak langkah pemindahan satu cakram dari batang awal ke batang tujuan
    TowerOfHanoi(n-1, aux_rod, to_rod, from_rod) # Pemanggilan rekursif kedua untuk memindahkan (n-1) cakram yang telah dipindahkan sebelumnya dari batang bantu ke batang tujuan


# Driver code
N = 3

# A, C, B are the name of rods
TowerOfHanoi(N, 'A', 'C', 'B')
```

# Mengurutkan Dalam Sebuah Stack

```python
# Mendefinisikan fungsi

def sortStack(input):
    tmpStack = [] # Membuat tumpukan sementara yang akan digunakan untuk menyimpan elemen-elemen yang sedang diurutkan
    while (len(input) > 0): # Melakukan perulangan selama tumpukan input tidak kosong

        # mengeluarkan elemen pertama
        tmp = input[-1] #  Mengambil elemen teratas dari tumpukan input
        input.pop() # Menghapus elemen teratas dari tumpukan input

        # ketika tumpukan sementara tidak kosong dan tumpukan bagian atas lebih kecil dari suhu

        while (len(tmpStack) > 0 and tmpStack[-1] < tmp): # Selama tumpukan sementara tidak kosong dan elemen teratasnya lebih kecil dari tmp

            # pop dari tumpukan sementara dan tambahkan ke tumpukan input
            input.append(tmpStack[-1]) # Mengambil elemen teratas tumpukan sementara dan menambahkannya ke tumpukan input
            tmpStack.pop() # Menghapus elemen teratas dari tumpukan sementara

        # tambahkan temp sementara dari tumpukan
        tmpStack.append(tmp) # Menambahkan tmp (elemen teratas dari tumpukan input) ke tumpukan sementara

    return tmpStack # Mengembalikan tumpukan sementara yang sudah diurutkan

# mendefinisikan fungsi
def sortArrayUsingStacks(arr, n):

    # tambahkan elemen array ke tumpukan
    input = [] # Membuat tumpukan input untuk menyimpan elemen-elemen array
    i = 0
    while ( i < n ): #  Melakukan perulangan untuk menambahkan elemen-elemen array ke tumpukan input
        input.append(arr[i]) # Menambahkan elemen array ke tumpukan input
        i = i + 1 # Menaikkan nilai i

    # Urutkan tumpukan sementara
    tmpStack = sortStack(input) # Memanggil fungsi sortStack untuk mengurutkan tumpukan input, dan hasilnya disimpan dalam tumpukan sementara tmpStack
    i = 0

    # Letakkan elemen tumpukan di arrp[]
    while (i < n): # Melakukan perulangan untuk menaruh elemen-elemen tumpukan sementara ke dalam array arr
        arr[i] = tmpStack[-1] # Menaruh elemen teratas tumpukan sementara ke dalam array arr
        tmpStack.pop() #  Menghapus elemen teratas dari tumpukan sementara
        i = i + 1 # Menaikkan nilai i

    return arr # Mengembalikan array yang sudah diurutkan

# Driver code
arr = [10, 5, 15, 45] # Mendefinisikan array arr yang akan diurutkan
n = len(arr) # Mendapatkan panjang array

arr = sortArrayUsingStacks(arr, n) # Memanggil fungsi sortArrayUsingStacks untuk mengurutkan array arr menggunakan tumpukan
i = 0

while (i < n): # Melakukan perulangan untuk mencetak elemen-elemen array yang sudah diurutkan
    print(arr[i] ,end= " ") # Mencetak elemen array tanpa newline
    i = i + 1 # Menaikkan nilai i
```
